```python
#MEASURE OF CENNTRAL TENDENCY
'''
1-mean(average):
       sum of observation/number of observation
2-median: calculated just by looking at the given data.
       even no of terms:take the mean of the two mid terms
       odd no of terms: take the mid term
3-mode:most frequently occuring element.

realtion: 3(median)=2(mean)+mode

'''
```

```python
#MEASURE OF SPREAD
'''
1-range=highest_value-lowest_value
        tells us the distribution of data.
2-IQR(interquartile range):
     quartile=quarter
     there are 4 quartile
     Q1,Q2,Q3,Q4 25,50,5,100 percentile respectively
     Qi=(psition of observation*100)/number of observation
   IQR=Q3-Q1
'''
```

```python
#5 NUMBER SUMMARY
'''
1-minimum
2-first quartile(Q1)
3-median(Q2)
4-third quartile(Q3)
5-maximum

outlier:
       alter the models performance leading to false prediction and poor efficieny.
       hinder the calculations of central tendancy.
  calcuate::
    upper fence=Q3+1.5(IQR)
    lower fence=Q1+1.5(IQR)
  Elements who are not in the range i.e not between upper fence and lower fence are outliers.
  (**after removing the outliers calculate the 5 number summary)
'''
```

```python
import numpy as np
arr=np.array([2,3,4,6,7,8,9,12,13,16,17,23,25,27,34,37,210])
arr
```

```python
q1=np.percentile(arr,25)
q1
q3=np.percentile(arr,75)
q3
iqr=q3-q1
iqr
```

```python

# upper and lower fence:
up = q3 + 1.5 * iqr
lw = q1 - 1.5 * iqr

up
```

```python
remove_out=[]
for i in arr:
  if i >=lw and i<=up:
    remove_out.append(i)
arr2=np.array(remove_out)
arr2
```

```python

import seaborn as sns
# with outliers
sns.boxplot(x=arr)
# without outliers
sns.boxplot(x=arr2)
```
