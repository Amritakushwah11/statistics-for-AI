 # BASIC PLOTS :
1-BAR GRAPH: firstly create a frequency table then plot the bars.

2-pie chart:represents the data using percentage
          created with the help of frequency table

frequency table: table that contains the frequency of each individual
example:
    apple          6
    mango          5
    banana         3
    grapes         4

3-line graph:created using bar graph .
           draw a dot on each bar and connect the dots ,erasing all the bars.
           helps in trend analysis and prediction.

4-o-gives: are similar to line plot but use cumulative frequency.

ONE WAY TABLE: ONE individual
TWO WAY TABLE: TWO individuals

5-dot plot: dots are ploted in specific range .

relative frequency table:frequency table created using a two way table.
                  data is stored and counted in percentage.
                  * used in data normalization.

6-histograms: data is stored into buckets or bins.
              it is similar to dot plot and bar graphs.
              bins are created using the range
              range=highest_value-lowest_value
              number of bins=range/random number
                                  random number=number of bins
