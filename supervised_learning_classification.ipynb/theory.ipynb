classification:

Logistic regression :(either 0 or 1)

linear regression is not suitable for classification task because:
1- the outliers can affect the best fit line.
2- result can be negative i.e. <0 or positive i.e. >1.
thus it is prefered to use logistic regression (sigmoid function).
