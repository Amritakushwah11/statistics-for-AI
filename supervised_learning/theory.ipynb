  used for continuous variables 

regression: (way to predict)

# LINEAR REGRESSION:
  creates a best fit line which helps in predicting the outcomes based on the given data.

  best fit line: straight line b/w x and y axis in the analysis plot.
  Y= MX+C             M=Slope  C=Intercept

  **RESIDUAL ERROR**:
    actual value - predicted value   
    calculated using 1-mean square value ( cost function)
                      2- gradient descent (creation of gradient descent curve using various values of theta1 and theta0 )

# performance matrices
 R**2: tells how good the model is (always positive)
 R² batata hai ki model data ko kitna achha explain kar raha hai.
 adjusted R**2: 
 Jab bhi tum model me extra variables add karte ho, R² hamesha increase hota hai, chahe wo variable useless ho.

👉 Is problem ko solve karta hai Adjusted R²

Definition:
Adjusted R² = R² ka improved version jo number of features (variables) ko consider karta hai.

# over fitting:(low bias high variance)
 model performs good while training but the performance decreases while testing
  Bias = model kitna galat assumption kar raha hai.
  Variance = model kitna sensitive hai data ke changes ke liye.

# under fitting:(high bias and high variance)
  model performs poor in both training as well as testing.

** ideal situation : (low bias and low variance) **

solution : use ridge and lasso regression.
           (include hyper parameter)
